```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
photos = np.load("SushiPhotos.npz")
```

```python
x = photos["x"]
img = x[13]
```

```python
plt.figure(figsize=(10, 10))

for i in range(18):
    plt.subplot(6, 6, i+1)
    center = (32, 32)
    angle = i*10
    scale = 1.0
    
    # Rotation
    mtx = cv2.getRotationMatrix2D(center, angle, scale)
    img2 = cv2.warpAffine(img, mtx, (64, 64))
    plt.imshow(img2)
    
plt.show()
```

```python
plt.figure(figsize=(10, 10))

for i in range(18):
    plt.subplot(6, 6, i+1)
    center = (32, 32)
    angle = i*10
    scale = 1.0
    mtx = cv2.getRotationMatrix2D(center, angle, scale)
    img2 = cv2.warpAffine(img, mtx, (64, 64))
    plt.imshow(img2)
    
    # Flip horizontal
    img3 = cv2.flip(img2, 1)
    plt.imshow(img3)
    
plt.show()
```

```python
plt.figure(figsize=(10, 10))

for i in range(18):
    plt.subplot(6, 6, i+1)
    center = (32, 32)
    angle = i*10
    scale = 1.0
    mtx = cv2.getRotationMatrix2D(center, angle, scale)
    img2 = cv2.warpAffine(img, mtx, (64, 64))

    # Blur
    average_square = (5, 5)
    img_blur = cv2.blur(img2, average_square)
    plt.imshow(img_blur)
    
plt.show()
```

```python
plt.figure(figsize=(10, 10))

for i in range(18):
    plt.subplot(6, 6, i+1)
    center = (32, 32)
    angle = i*10
    scale = 1.0
    mtx = cv2.getRotationMatrix2D(center, angle, scale)
    img2 = cv2.warpAffine(img, mtx, (64, 64))
    
    # Contrast
    contrast = 128
    factor = (259 *(contrast + 255)) / (255 *(259 - contrast))
    newImage = np.array(img2, dtype = 'float64')
    newImage = np.clip((newImage[:,:,:] - 128) * factor + 128, 0, 255)
    newImage = np.array(newImage, dtype = 'uint8')
    plt.imshow(newImage)
    
plt.show()
```
